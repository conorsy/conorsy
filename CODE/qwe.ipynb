```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

NOAA Climate Data Record (CDR) of AVHRR Normalized Difference Vegetation Index (NDVI), Version 5

```python
NDVI_BIG = xr.open_dataset("../DATA/BIG.nc", decode_times=False)
NDVI_SUBSET = xr.open_dataset("../DATA/SUBSET.nc", decode_times=False)
```
